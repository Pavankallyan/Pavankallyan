# **Topic- University Student Information System**

# **Phase 1: Planning**

1. Define the Scope

The University Student Information System will manage and organize data related to:
Students: Personal details, enrollment information, and academic records.
Courses: Course details, credits, and schedules.
Faculty: Faculty details, courses taught, and departments.
Departments: Department details and their associated faculty and courses.
Enrollment: Tracks which students are enrolled in which courses, along with semester and grades.

The project should allow:

CRUD operations for all entities (students, courses, faculty, departments, and enrollment).

Retrieval of specific information through queries (e.g., student enrollment, faculty schedules, course details).

Optional: Role-based access for admin, faculty, and students.

2. Define Objectives

The main objective of the project is to design and implement a relational database system that:

Efficiently stores and retrieves data related to students, courses, faculty, and enrollment.

Ensures data integrity and minimizes redundancy.

Provides answers to specific queries, such as:

How many students are enrolled in a specific course?
Which courses are taught by a particular faculty member?
What is the total enrollment in each department?
What grades did a student achieve in a specific semester?

3. Functional Requirements

The system must fulfill the following functional requirements:
CRUD Operations:

Create, Read, Update, and Delete records for all entities (students, courses, faculty, departments, and enrollment).

Queries:

Retrieve data such as:

List of students enrolled in a course.

Faculty teaching a specific course.

Total enrollment in each department.

Grades of a student in a semester.

Data Integrity:

Use primary keys, foreign keys, and constraints to ensure data consistency.

Normalization:

Ensure the database is normalized to at least 3rd Normal Form (3NF).

4. Set Objectives for Queries

The database should be able to answer the following questions:

How many students are enrolled in a specific course?

Which courses are taught by a specific faculty member?

What is the total enrollment in each department?

What grades did a student achieve in a specific semester?

Which students are enrolled in courses taught by a specific faculty member?

Step 2: Define Relationships

The relationships between the entities are as follows:

Students and Departments:

A student belongs to one department.

A department can have many students.

Relationship: One-to-Many (1:N).

Courses and Departments:

A course belongs to one department.

A department can offer many courses.

Relationship: One-to-Many (1:N).

Faculty and Departments:

A faculty member belongs to one department.

A department can have many faculty members.

Relationship: One-to-Many (1:N).

Students and Courses (via Enrollment):

A student can enroll in many courses.

A course can have many students enrolled.

Relationship: Many-to-Many (M:N), resolved through the Enrollment associative entity.

Faculty and Courses:

A faculty member can teach many courses.

A course is taught by one faculty member.

Relationship: One-to-Many (1:N).

Step 3: Add Cardinalities

The cardinalities for the relationships are:

Students and Departments: 1 department can have many students, but each student belongs to 1 department.

Courses and Departments: 1 department can offer many courses, but each course belongs to 1 department.

Faculty and Departments: 1 department can have many faculty members, but each faculty member belongs to 1 department.

Students and Courses (via Enrollment): Many students can enroll in many courses.
Faculty and Courses: 1 faculty member can teach many courses, but each course is taught by 1 faculty member.

Step 4: Design the ER Diagram


Entities: Represented as rectangles.

Relationships: Represented as diamonds connecting entities.

Primary Keys: Underlined attributes.

Foreign Keys: Marked as attributes that reference primary keys in other entities.
